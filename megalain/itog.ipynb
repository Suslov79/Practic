    - Обзор данных.

Каждый звонок в тарифах компании округляется отдельно: даже если он длился всего 1 секунду, будет засчитан как 1 минута.

```python
# Функция округления в большую сторону и изменения типа до целочисленного
def round_int(df, column):
    df[column] = df[column].apply(mt.ceil)
    df[column].astype(int)
```

## Исследовательский анализ данных:

### Количество сделанных звонков и израсходованных минут разговора по месяцам

```python
# группируем данные с применением агригирующих функций по id пользователей
calls_month= calls.groupby(['user_id','month']).agg({'duration': ['sum', 'count']})
calls_month.head(20)                        
```

### Количество отправленных сообщений по месяцам

```python
# группируем данные с применением агригирующих функций по id пользователей
messages_month= messages.groupby(['user_id','month']).agg({'id': ['count']})
messages_month.head(20)
```

### Oбъем израсходованного интернет-трафика по месяцам

```python
# группируем данные с применением агригирующих функций по id пользователей
internet_month= internet.groupby(['user_id','month']).agg({'mb_used': ['sum']})
internet_month.head(20)
```

### Помесячная выручка с каждого пользователя 

```python
# для соединения трех таблиц сбросим индексы (id пользователя пропишется в каждой строке- по нему пройдет соединение)
calls_month.reset_index()
messages_month.reset_index()
internet_month.reset_index()
```

```python
# формируем итоговую таблицу
total = internet_month.merge (calls_month,on=['user_id', 'month'], how='outer') 
total = total.merge(messages_month,on=['user_id', 'month'], how='outer')
total.sample(10,random_state=10)
```

```python
total.reset_index(inplace=True)
total.info()  # получение общей информации о данных
```

Для веб-трафика отдельные сессии не считаются. Вместо этого общая сумма за месяц округляется в бо́льшую сторону. Если абонент использует 1025 мегабайт в этом месяце, с него возьмут плату за 2 гигабайта.

```python
# Cкорректируем наименования столбцов и приведем объем интернеит трафика к Гигабайтам. 
# Затем округлим сзначения до целых в большую сторону
total.columns = ['user_id', 'month','total_gb_used', 'total_calls_duration', 'total_calls', 'total_messages']
total=total.fillna(0)
total['total_gb_used'] = total['total_gb_used']/1024
round_int(total,'total_gb_used')
total.sample(10,random_state=10)
```

```python
total_users=total.merge(users, on=['user_id'], how='outer') # добавим данные о пользователях
total_users
```

Неиспользованные за предыдущий месяц звонки, смс, интернет на следующий месяц не переносятся.Вычтем бесплатный лимит из суммарного количества звонков, сообщений и интернет-трафика

```python
total_users['gb_for_payment'] =(total_users['total_gb_used']-(total_users['mb_per_month_included']/1024))
total_users['minutes_for_payment'] =(total_users['total_calls_duration']-total_users['minutes_included'])
total_users['messages_for_payment'] =(total_users['total_messages']-total_users['messages_included'])
```

```python
total_users['gb_for_payment'].describe()
```

```python
total_users['minutes_for_payment'].describe()
```

```python
total_users['messages_for_payment'].describe()
```

Получилась часть аббонентов не расходует трафик полностью, что привело к отрицательным значениям. Такого быть не может , поэтому заменим такие данные на "0"

```python
total_users.loc[total_users['minutes_for_payment'] < 0, 'minutes_for_payment'] = 0
```

```python
total_users.loc[total_users['messages_for_payment'] < 0, 'messages_for_payment'] = 0
```

```python
total_users.loc[total_users['gb_for_payment'] < 0, 'gb_for_payment'] = 0
```

```python
total_users['gb_for_payment'].describe()
```

```python
total_users['minutes_for_payment'].describe()
```

```python
total_users['messages_for_payment'].describe()
```

Перерасход умножим на значение из тарифного плана и прибавим абонентскую плату, соответствующую тарифному плану. Если  у клиента "0" перерасхода - то он платит только сумму в соответсвии с тарифным планом.

```python
total_users['total_payment'] = (total_users['minutes_for_payment'] * total_users['rub_per_minute'] +
                               total_users['messages_for_payment'] * total_users['rub_per_message'] +
                               total_users['gb_for_payment'] * total_users['rub_per_gb'])+total_users['rub_monthly_fee']

total_users  # просмотр после преобразования 
```

```python
# можем получить итоговые оплаты от каждого пользователя в месяц
revenue_per_users= total_users[['user_id','month','total_payment']]
revenue_per_users
```

```python
display ('Суммарная выручка за рассматриваемый период {} рублей'.format(sum(revenue_per_users['total_payment'])))
```

## Проверка гипотез

### Cредняя выручка пользователей тарифов «Ultra» и «Smart» различаются

В нашем распряжении пара совокупностей.Они не зависят друг от друга. Проверим гипотезу о равенстве средних двух совокупностей. 
Учтем что дисперсии в них различны. Сравним средние t-тестом методом ttest_ind()

Сформулируем нулевую гипотезу H₀ :
- Средняя выручка двух трарифов одинакова.
- Альтернативная гипотеза H₁ - они различны 

```python
sample_1 = ultra_total['total_payment']
sample_2 = smart_total['total_payment']
alpha = .05
results = st.ttest_ind(sample_1, sample_2, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу.')
else:
    print('Не получилось отвергнуть нулевую гипотезу.')
```

Вероятность получить равные значения средней выручки для двух тарифов ничтожно мала. Что дает нам веские основания отвергнуть нулевую гипотезу.  

### Cредняя выручка пользователей из Москвы отличается от выручки пользователей из других регионов.

В нашем распряжении ещё одна пара совокупностей.Они так же не зависят друг от друга. 
Проверим гипотезы о равенстве средних двух совокупностей. Учтем что дисперсии в них различны. 
Сравним средние t-тестом методом ttest_ind()

Сформулируем нулевую гипотезу H₀ :
- Средняя выручка пользователей из Москвы и других городов одинакова.
- Альтернативная гипотеза H₁ - они различны 

```python
moscow_total= final.query('city == "Москва"')   # формируем таблицу пользователей только из Москвы
moscow_total
```

```python
moscow_total.describe() # описание данных
```

```python
city_total= final.query('city != "Москва"')  # формируем таблицу пользователей кроме Москвы
city_total
```

```python
city_total.describe() # описание данных
```

```python
sample_1 = moscow_total['total_payment']
sample_2 = city_total['total_payment']
alpha = .05
results = st.ttest_ind(sample_1, sample_2, equal_var=False)
print('p-значение:', results.pvalue)
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу.')
else:
    print('Не получилось отвергнуть нулевую гипотезу.')
```

Вероятность получить одинаковые значения выше чем установленный нами порог уровня статистической значимости. Занчит оснований отвергуть нулевую гипотезу у нас нет.
