### Шаг 2.6. Категоризация дохода.

Для ускорения принятия решения о кредитных рисках можем ввести "категории клиентов" по уровню их доходов. Это нам поможет не вглядываться в цифры а быстрее определять платежеспособность клиента.

    0–30000 — 'E';
    30001–50000 — 'D';
    50001–200000 — 'C';
    200001–1000000 — 'B';
    1000001 и выше — 'A'.
    Например, кредитополучателю с доходом 25000 можно назначить категорию 'E', а клиенту, получающему 235000, — 'B'.

```python
def income_category(income):  # фунция автоматически проверит условия для каждого клиента и присвоит ему необходимую категорию
    try:
        if income <= 30000 :
            return 'E';
        if income <= 50000 :
            return 'D';
        if income <= 200000 :
            return 'C';
        if income <= 1000000 :
            return 'B';
        if income > 1000000 :
            return 'A';
    except:
        return 'уточнить доход'
df['total_income_category']=df['total_income'].apply (income_category)
df
```

### Шаг 2.7. Категоризация целей кредита.

Каждые клиент может по своему описывать цели кредита, но в целом все они попадают в достаточно узкий круг не из десятков, а всего из нескольких необходимых вещей. Так же можем упростить работу с данным сегментом.

Создадим функцию, которая на основании данных из столбца purpose сформирует новый столбец purpose_category, в который войдут следующие категории:

    ‘операции с автомобилем’,
    ‘операции с недвижимостью’,
    ‘проведение свадьбы’,
    ‘получение образования’.

Например, если в столбце purpose находится подстрока 'на покупку автомобиля', то в столбце purpose_category должна появиться строка 'операции с автомобилем'.

```python
def purpose_category(purpose):    # фунция автоматически проверит категоию для каждого клиента и присвоит ему необходимое значение
    try:
        if  "авто" in purpose:
            return 'операции с автомобилем';
        if  "недвижимо" in purpose or "жил" in purpose:
            return 'операции с недвижимостью';
        if  "свадьб" in purpose:
            return 'проведение свадьбы';
        if  "образован" in purpose:
            return 'получение образования';
    except:
        return 'уточнить цель кредита'
df['purpose_category']=df['purpose'].apply(purpose_category)
df
```
